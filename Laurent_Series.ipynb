# Definition
An annulus  in the complex plane is the open region between two concentric circles centered at some point $z_0 \in \mathbb{C}.$

Formally, for $0 \leq R_1 < R_2 \leq \infty$, the annulus centered at $z_0$​ is defined as:

$$A(z_0; R_1, R_2) = \{ z \in \mathbb{C} \; : \; R_1 < |z - z_0| < R_2 \}$$

## Intuition

If $R_1 = 0$, the annulus reduces to the open disk of radius $R_2$​.

If $R_2 = \infty$, the annulus becomes the exterior region $|z-z_0| > R_1​.$

Otherwise, it’s a “ring-shaped” region that excludes the inner circle $|z-z_0|=R_1$ and outer circle $|z-z_0|=R_2$​.

## Significance

The Laurent series converges precisely in an annulus.

 The inner radius $R_1$​ is determined by the nearest singularity inside,

 The outer radius $R_2$​ is determined by the nearest singularity outside.

 So, the annulus captures exactly the region where the series is valid.


```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters for annulus
R1, R2 = 1, 3
z0 = 0  # center at origin

theta = np.linspace(0, 2*np.pi, 400)
x_inner = R1 * np.cos(theta)
y_inner = R1 * np.sin(theta)
x_outer = R2 * np.cos(theta)
y_outer = R2 * np.sin(theta)

fig, ax = plt.subplots(figsize=(7,7))

# Fill annulus region
ax.fill_betweenx(y_outer, x_inner, x_outer, color="lightblue", alpha=0.5, label="Annulus region")

# Draw inner and outer circles
ax.plot(x_inner, y_inner, 'r-', linewidth=1.5, label=r"$|z-z_0| = R_1$ (inner boundary)")
ax.plot(x_outer, y_outer, 'b-', linewidth=1.5, label=r"$|z-z_0| = R_2$ (outer boundary)")

# Mark center z0
ax.plot(0,0,'ko')
ax.text(0.1,0.1, r"$z_0$", fontsize=12, color='black')

# Annotate R1 and R2
ax.annotate(r"$R_1$", xy=(R1/np.sqrt(2), R1/np.sqrt(2)), xytext=(1.3,0.5),
            arrowprops=dict(arrowstyle="->", color="red"))
ax.annotate(r"$R_2$", xy=(R2/np.sqrt(2), R2/np.sqrt(2)), xytext=(3.3,1.5),
            arrowprops=dict(arrowstyle="->", color="blue"))

# Cosmetics
ax.set_xlim(-3.8,3.8)
ax.set_ylim(-3.8,3.8)
ax.set_aspect('equal')
ax.set_xlabel("Re(z)")
ax.set_ylabel("Im(z)")
ax.set_title("General Annulus Region in Complex Plane")
ax.legend(loc="upper right")
ax.grid(True, linestyle=":", alpha=0.6)

plt.show()
```

Here’s the **geometric illustration of a general annulus** in the complex plane:

- The **center** is at  $z_0$.
- The **inner boundary** is the circle $ |z - z_0| = R_1 $ (red).
- The **outer boundary** is the circle  $|z - z_0| = R_2 $ (blue).
- The shaded **ring-shaped region** between them is the annulus:

$$A(z_0; R_1, R_2) = \{ z \in \mathbb{C} : R_1 < |z - z_0| < R_2 \}.$$

This is exactly the domain where a **Laurent series** converges.

To deepen the intuition behind the Laurent series, it helps to visualize the geometry of its domain—specifically, the annulus. Imagine a donut-shaped region in the complex plane, centered around a singularity. This annulus lies between two concentric circles: the inner boundary excludes the singularity, while the outer boundary marks the limit of convergence. Within this ring, the Laurent series converges beautifully, weaving together both positive and negative powers of $𝑧$. The regular part (positive powers) behaves like a Taylor series, capturing smooth variations as you move outward. The principal part (negative powers), however, dives inward, decoding the behavior as you approach the singularity. By sketching this annular domain and labeling the zones of convergence, we see how the Laurent series acts like a bridge—linking the calm exterior with the swirling core. This geometric lens transforms abstraction into clarity, showing not just what the series does, but where and why it works.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure
fig, ax = plt.subplots(figsize=(6,6))

# Parameters for the annulus
r_inner = 1   # inner radius
r_outer = 3   # outer radius

# Plot annulus (region of convergence)
theta = np.linspace(0, 2*np.pi, 500)
x_outer = r_outer * np.cos(theta)
y_outer = r_outer * np.sin(theta)
x_inner = r_inner * np.cos(theta)
y_inner = r_inner * np.sin(theta)

# Fill annulus
ax.fill(x_outer, y_outer, color='lightblue', alpha=0.5, label='Region of Convergence')
ax.fill(x_inner, y_inner, color='white')

# Mark the center (z0)
ax.plot(0, 0, 'ro')
ax.text(0.1, 0.1, r"$z_0$", fontsize=14, color="red")

# Mark circles r and R
circle_r = plt.Circle((0,0), r_inner, color='blue', linestyle='--', fill=False, label=r'$|z-z_0|=r$')
circle_R = plt.Circle((0,0), r_outer, color='blue', linestyle='-', fill=False, label=r'$|z-z_0|=R$')
ax.add_patch(circle_r)
ax.add_patch(circle_R)

# Formatting
ax.set_aspect('equal')
ax.set_xlim(-3.5,3.5)
ax.set_ylim(-3.5,3.5)
ax.set_title("Annulus of Convergence of Laurent Series", fontsize=14)
ax.legend(loc="upper right")

plt.show()
```

Here’s the geometric intuition for a Laurent series expansion:

* The red dot is the expansion point $z_0$
* The dashed inner circle ($|z-z_0| =r$) marks where the series cannot cross inward (due to a singularity inside).
* The solid outer circle ($|z-z_0|=R$) marks where the series cannot extend outward (due to another singularity or boundary).
* The blue shaded annulus is the region where the Laurent series converges.

So, unlike a Taylor series (which converges only in a disk), a Laurent series converges in a ring-shaped domain.

I marked the two expansions of $f(z)=\dfrac{1}{z(z-1)}$ ​ about $z_0=0, z=1$ on two diagrams you can see below.

```python
import matplotlib.pyplot as plt
import numpy as np

# Option A with geometric indications added clearly

theta = np.linspace(0, 2*np.pi, 400)
r = 1.0

fig, axes = plt.subplots(1, 2, figsize=(16,7))

# ============================================================
# LEFT PANEL — Expansion about z0 = 0
# ============================================================
ax = axes[0]

# Geometry
x = r*np.cos(theta)
y = r*np.sin(theta)

ax.fill(x, y, color="lightblue", alpha=0.35)
ax.plot(x, y, color="blue", linewidth=1.8)

# Singularities
ax.plot(0,0,'ro', markersize=8)
ax.text(0.05,0.1, r"$z_0=0$", color="red", fontsize=12)

ax.plot(1,0,'mo', markersize=8)
ax.text(1.05,0.05, r"$z=1$", color="purple", fontsize=12)

# Indication arrows
ax.annotate(
    "Outer boundary\n(nearest singularity)",
    xy=(1,0), xytext=(1.3,0.8),
    arrowprops=dict(arrowstyle="->", linewidth=1.2),
    fontsize=11, bbox=dict(facecolor="white", alpha=0.9)
)

ax.annotate(
    "Inner boundary\n(pole at center)",
    xy=(0,0), xytext=(-1.4,-0.8),
    arrowprops=dict(arrowstyle="->", linewidth=1.2),
    fontsize=11, bbox=dict(facecolor="white", alpha=0.9)
)

ax.annotate(
    "Domain of convergence:\n$0<|z|<1$",
    xy=(0.4,0.4), xytext=(-1.55,1.25),
    arrowprops=dict(arrowstyle="->"),
    fontsize=11, bbox=dict(facecolor="white", alpha=0.9)
)

# Cosmetics
ax.set_xlim(-1.7,1.7)
ax.set_ylim(-1.7,1.7)
ax.set_aspect("equal")
ax.grid(True, linestyle=":", alpha=0.6)
ax.set_title("Laurent Expansion about $z_0 = 0$")
ax.set_xlabel("Re(z)")
ax.set_ylabel("Im(z)")


# ============================================================
# RIGHT PANEL — Expansion about z0 = 1
# ============================================================
ax = axes[1]

x1 = 1 + r*np.cos(theta)
y1 = r*np.sin(theta)

ax.fill(x1, y1, color="lightblue", alpha=0.35)
ax.plot(x1, y1, color="blue", linewidth=1.8)

# Singularities
ax.plot(1,0,'mo', markersize=8)
ax.text(1.07,0.1, r"$z_0=1$", fontsize=12, color="purple")

ax.plot(0,0,'ro', markersize=8)
ax.text(0.1,0.08, r"$z=0$", fontsize=12, color="red")

# Arrows + indications
ax.annotate(
    "Outer boundary\n(nearest singularity)",
    xy=(0,0), xytext=(-1.4,0.8),
    arrowprops=dict(arrowstyle="->", linewidth=1.2),
    fontsize=11, bbox=dict(facecolor="white", alpha=0.9)
)

ax.annotate(
    "Inner boundary\n(pole at center)",
    xy=(1,0), xytext=(2.3,-0.8),
    arrowprops=dict(arrowstyle="->", linewidth=1.2),
    fontsize=11, bbox=dict(facecolor="white", alpha=0.9)
)

ax.annotate(
    "Domain of convergence:\n$0<|z-1|<1$",
    xy=(1.6,0.4), xytext=(2.15,1.25),
    arrowprops=dict(arrowstyle="->"),
    fontsize=11, bbox=dict(facecolor="white", alpha=0.9)
)

# Cosmetics
ax.set_xlim(-1.7,3.7)
ax.set_ylim(-1.7,1.7)
ax.set_aspect("equal")
ax.grid(True, linestyle=":", alpha=0.6)
ax.set_title("Laurent Expansion about $z_0 = 1$")
ax.set_xlabel("Re(z)")

fig.suptitle(r"Laurent Expansions of $f(z)=\frac{1}{z(z-1)}$ about $z_0=0$ and $z_0=1$", fontsize=16)

plt.tight_layout()
plt.show()
```

## ✔ Inner Boundary

Shows why the inner radius is zero:

* Because the center $𝑧_0$ is itself a pole.

* The Laurant expansion domain therefore starts as a punctured disk.

## ✔ Outer Boundary

Shows that the next nearest singularity determines the outer radius:

* At $∣z∣=1$ for expansion about $z=0$

* At $∣z−1∣=1$ for expansion about $𝑧_0=1$

✔ Domain of Convergence (clearly indicated)

The clean arrow and boxed text describe exactly where the Laurent series converges.

```python
# Animation + numerical verification + residue-theorem demo for f(z)=1/(z(z-1))
# This code is written to run in Colab (or any Jupyter environment).
# It produces:
# 1) an animation (saved to /content/laurent_animation.gif) showing a contour radius moving and indicating
#    whether the pole at z=1 is inside the contour (hence whether the Laurent expansion about 0
#    contains the contribution from z=1)
# 2) numerical computation of Laurent coefficients a_n via the integral formula for two radii (rho=0.5 and rho=2)
# 3) residue-theorem evaluation of a real integral I = ∫_0^{2π} dθ / (2 + cos θ) and numeric check
#
# Download link will be shown after the animation is saved.
#
# To run in Colab: paste the whole cell, run it. If animation saving requires `pillow` it should already be present.
# If you want mp4 instead, install ffmpeg in Colab and change the writer accordingly.

# =============================================
# Laurent Series Animation + Numerical Check
# for f(z) = 1 / (z (z-1))
# =============================================

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
import math
import pandas as pd
from IPython.display import display

# --- Part A: Animation ---
fig, ax = plt.subplots(figsize=(7,7))   # bigger figure
ax.set_xlim(-3.2, 3.2)
ax.set_ylim(-3.2, 3.2)
ax.set_aspect('equal')
ax.set_xlabel('Re(z)', fontsize=12)
ax.set_ylabel('Im(z)', fontsize=12)
ax.grid(True, linestyle=':', linewidth=0.6, alpha=0.6)
ax.set_title('Moving contour around 0 and inclusion of pole at z=1', fontsize=13)

# static markers
ax.plot(0, 0, 'o', color="black")   # center
ax.plot(1, 0, 'o', color="green")   # pole at z=1
ax.text(0.05, 0.05, '0', fontsize=11)
ax.text(1.05, 0.05, '1', fontsize=11)

contour_line, = ax.plot([], [], lw=2, color="red")
status_text = ax.text(0, 2.6, '', fontsize=10, ha='center',
                      bbox=dict(facecolor='white', alpha=0.8))

radii = np.linspace(0.2, 2.4, 120)

def init():
    contour_line.set_data([], [])
    status_text.set_text('')
    return contour_line, status_text

def animate(i):
    rho = radii[i]
    theta = np.linspace(0, 2*np.pi, 400)
    x = rho * np.cos(theta)
    y = rho * np.sin(theta)
    contour_line.set_data(x, y)
    inside = (rho > 1 - 1e-9)  # if radius exceeds 1, the pole at z=1 is inside
    if inside:
        status = (f'Contour radius = {rho:.2f} → pole at z=1 is INSIDE contour\n'
                  f'Laurent about 0 will include contribution from z=1')
    else:
        status = (f'Contour radius = {rho:.2f} → pole at z=1 is OUTSIDE contour\n'
                  f'Laurent about 0 has principal part from z=0 and analytic tail from 1/(z-1)')
    status_text.set_text(status)
    return contour_line, status_text

anim = FuncAnimation(fig, animate, frames=len(radii),
                     init_func=init, blit=True, interval=80)

# Save animation
gif_path = "/content/laurent_animation.gif"
writer = PillowWriter(fps=12)
anim.save(gif_path, writer=writer)
plt.close(fig)
print("Animation saved to:", gif_path)


# --- Part B: Numerical Laurent coefficients ---
def f(z):
    return 1.0/(z*(z-1.0))

def compute_a_n(rho, n_vals, Nint=4000):
    t = np.linspace(0, 2*np.pi, Nint, endpoint=False)
    z = rho * np.exp(1j * t)
    ft = f(z)
    results = {}
    prefactor = 1.0/(2*np.pi)
    for n in n_vals:
        integrand = ft * np.exp(-1j * n * t)
        integral = np.trapz(integrand, t)
        a_n = prefactor * (rho**(-n)) * integral
        results[n] = a_n
    return results

n_list = list(range(-3, 4))
rho1, rho2 = 0.5, 2.0
coeffs_rho1 = compute_a_n(rho1, n_list)
coeffs_rho2 = compute_a_n(rho2, n_list)

def coeffs_to_df(coeffs, rho):
    df = pd.DataFrame({
        "n": list(coeffs.keys()),
        f"a_n (rho={rho}) [Real]": [round(v.real, 6) for v in coeffs.values()],
        f"a_n (rho={rho}) [Imag]": [f"{v.imag:.2e}" for v in coeffs.values()]
    })
    return df

df_rho1 = coeffs_to_df(coeffs_rho1, rho1)
df_rho2 = coeffs_to_df(coeffs_rho2, rho2)

print("\n### Numerical Laurent Coefficients (Computed) ###")
display(df_rho1)
display(df_rho2)

# Theoretical expectations
theory_data = {
    "Case": ["|z|<1 expansion", "|z|>1 expansion"],
    "Expected key coefficients": [
        "a_{-1} = -1 ; a_n = -1 for n ≥ 0",
        "a_{-2}=1 ; a_{-3}=1 ; a_{-4}=1 ; ..."
    ]
}
df_theory = pd.DataFrame(theory_data)
print("### Theoretical Expectations ###")
display(df_theory)


# --- Part C: Residue-theorem demonstration ---
coeffs = [1, 4, 1]
roots = np.roots(coeffs)

def residue_rational_num_over_quad(z0):
    denom_deriv = 2*z0 + 4
    return 2.0 / (1j * denom_deriv)

inside_roots = []
residues = []
for r in roots:
    if abs(r) < 1 + 1e-12:
        inside_roots.append(r)
        residues.append(residue_rational_num_over_quad(r))

sum_residues = sum(residues)
I_by_residue = 2j * math.pi * sum_residues
I_theory = 2*math.pi / math.sqrt(3)

# Table for poles and residues
residue_df = pd.DataFrame({
    "Root": [str(r) for r in roots],
    "Inside Unit Circle?": [abs(r)<1 for r in roots],
    "Residue": [f"{res.real:.6f} {res.imag:+.6f}i" if abs(r)<1
                else "-" for r,res in zip(roots, [residue_rational_num_over_quad(r) for r in roots])]
})

# Table for integral evaluation
integral_df = pd.DataFrame({
    "Method": ["Residue Theorem", "Theoretical (2π/√3)", "Absolute Error"],
    "Value": [I_by_residue, I_theory, abs(I_by_residue - I_theory)]
})

print("### Residue Theorem Poles and Residues ###")
display(residue_df)

print("### Integral Evaluation ###")
display(integral_df)

print("\nDownload the animation from:", gif_path)
```

An animation that you can run in Colab showing a contour centered at $0$ whose radius grows from $0.2 $ to $2.4$. The animation highlights when the pole at $z=1$ becomes inside the contour and explains the change in which Laurent terms remain or appear.

A numeric verification of Laurent coefficients $a_n$​ computed from the integral formula



$$
a_n=\frac{1}{2\pi i}\oint_\gamma \frac{f(\zeta)}{(\zeta-z_0)^{n+1}}\,d\zeta
$$

(with $z_0=0$ and $\gamma$ the circle $|\zeta|=\rho$). I computed coefficients for $n=-3,\dots,3$, at two radius

-   $\rho=0.5$ (inside unit disk): numerics match the expected expansion $\dfrac{1}{z}-1-z-z^2-\cdots$. In particular $a_{-1}\approx -1,\; a_0\approx -1,\; a_1\approx -1,\dots$

-   $\rho=2.0$ (outside unit disk): numerics match the expected exterior expansion $\displaystyle \dfrac{1}{z^2}+\dfrac{1}{z^3}+\cdots$ (so coefficients like $a_{-2}\approx 1,\ a_{-3}\approx 1,\dots$)

The printed numeric outputs are in the execution above (small rounding errors only). A residue-theorem worked example. I evaluated

$$I=\int_{0}^{2\pi}\frac{d\theta}{2+\cos\theta}$$​by mapping $z=e^{i\theta}$, converting the integral to a complex contour integral on $∣z∣=1$, finding the poles of the rational integrand inside the unit circle, summing residues, and applying $\pi$ times the sum. The residue computation gives

$$
I=\frac{2\pi}{\sqrt{3}} \approx 3.627598728468436
$$

and the numeric residue-sum agreed to machine precision.
