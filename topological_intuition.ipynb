# **Mathematical Perspective of the Plots (Topological Intuition)**

We are visualizing a **topological idea** using the familiar geometry of the plane $\mathbb{R}^2$. The pictures do **not** claim that every topology looks like circles in the plane; rather, they give the standard intuition that **open sets are built from neighborhoods**.

## **1. The Underlying Set $X$**

In the plots, we first fix a **finite set of points**

$$
X=\{x_1,x_2,\dots,x_6\}\subset \mathbb{R}^2.
$$

Each $x_i$ is a point in the plane, plotted and labeled.  

>**Important Viewpoint:**  
We are treating $X$ as an “abstract set,” but we *embed it* into $\mathbb{R}^2$ so we can draw it.


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# A simple "topological picture" in R^2:
# - X: a set of points (a finite subset of the plane)
# - U, V: example open sets, visualized as unions of open balls

# Finite subset X ⊂ R^2
X = np.array([
    [0.15, 0.20],
    [0.35, 0.75],
    [0.55, 0.30],
    [0.70, 0.80],
    [0.82, 0.45],
    [0.25, 0.45],
])

# "Open balls" whose unions we call U and V
balls_U = [((0.25, 0.45), 0.22), ((0.55, 0.30), 0.20)]
balls_V = [((0.70, 0.80), 0.22), ((0.82, 0.45), 0.18)]

fig = plt.figure(figsize=(7, 6))
ax = plt.gca()

# Plot points of X
ax.scatter(X[:, 0], X[:, 1], s=60)
for i, (x, y) in enumerate(X, start=1):
    ax.text(x + 0.015, y + 0.015, f"x{i}", fontsize=10)

# Draw U as a union of open balls
for (cx, cy), r in balls_U:
    ax.add_patch(Circle((cx, cy), r, fill=True, alpha=0.18))
    ax.add_patch(Circle((cx, cy), r, fill=False, linewidth=1.5))
ax.text(0.12, 0.93, "U = union of these open balls", fontsize=11)

# Draw V as a union of open balls
for (cx, cy), r in balls_V:
    ax.add_patch(Circle((cx, cy), r, fill=True, alpha=0.18))
    ax.add_patch(Circle((cx, cy), r, fill=False, linewidth=1.5))
ax.text(0.12, 0.88, "V = union of these open balls", fontsize=11)

ax.set_title("A visual intuition for a topological space: (X, τ) from open neighborhoods in R²")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect("equal", adjustable="box")
ax.grid(True, linewidth=0.5, alpha=0.5)

plt.show()
#______________________________________________________________________________

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Patch

# Finite subset X ⊂ R^2
X = np.array([
    [0.15, 0.20],
    [0.35, 0.75],
    [0.55, 0.30],
    [0.70, 0.80],
    [0.82, 0.45],
    [0.25, 0.45],
])

# "Open balls" whose unions define U and V
balls_U = [((0.25, 0.45), 0.22), ((0.55, 0.30), 0.20)]
balls_V = [((0.70, 0.80), 0.22), ((0.82, 0.45), 0.18)]

def draw_panel(ax, balls, title):
    # Plot points of X
    ax.scatter(X[:, 0], X[:, 1], s=60)
    for i, (x, y) in enumerate(X, start=1):
        ax.text(x + 0.015, y + 0.015, f"x{i}", fontsize=10)

    # Draw the open balls
    for (cx, cy), r in balls:
        ax.add_patch(Circle((cx, cy), r, fill=True, alpha=0.18))
        ax.add_patch(Circle((cx, cy), r, fill=False, linewidth=1.5))

    ax.set_title(title)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect("equal", adjustable="box")
    ax.grid(True, linewidth=0.5, alpha=0.5)

# One window, two frames
fig, axes = plt.subplots(1, 2, figsize=(12, 5), constrained_layout=True)

draw_panel(axes[0], balls_U, "Open set U = union of open balls")
draw_panel(axes[1], balls_V, "Open set V = union of open balls")

# Legend note (same on both frames)
legend_handles = [Patch(alpha=0.18, label="open ball (neighborhood)")]
for ax in axes:
    ax.legend(handles=legend_handles, loc="lower left", frameon=True)

fig.suptitle("Same Topological Picture, Separated into Two frames: U and V", fontsize=14)
plt.show()
```

## **2. Open Balls as Neighborhoods**

In $\mathbb{R}^2$ with the Euclidean distance $d$, an **open ball** centered at $a\in \mathbb{R}^2$ with radius $r>0$ is

$$
B(a,r)=\{y\in \mathbb{R}^2:\ d(y,a)<r\}.
$$

> Geometrically, this is the interior of a circle (boundary not included).  
In the code, each “ball” is drawn as a circle with a lightly shaded interior, representing $B(a,r)$.


## **3. What the Picture is Encoding: Open Sets from Neighborhoods**

The plots define two sets, $U$ and $V$, each formed by taking a **union of open balls**.

### **The Set $U$**
The **left frame** shows:

$$
U = B(a_1,r_1)\ \cup\ B(a_2,r_2).
$$

So $U$ is open in $\mathbb{R}^2$ because it is a union of open sets.

### **The set $V$**
The **right frame** shows:

$$
V = B(b_1,s_1)\ \cup\ B(b_2,s_2).
$$

Similarly, $V$ is open in $\mathbb{R}^2$.

**Main Topological Message:**
> In metric spaces, open sets can be constructed as unions of open balls (neighborhoods).



## **4. Connecting to the Topology $\tau$**

Let $(\mathbb{R}^2,\tau)$ be the plane with its **standard topology**, generated by open balls.

Then:
- $(U \in \tau)$,
- $(V \in \tau)$,

because both are unions of open balls.

The pictures illustrate the “**axioms of topology**” in the background:
1. $\emptyset$ and $\mathbb{R}^2$ are open.
2. Any union of open sets is open.
3. Any finite intersection of open sets is open.



## **5. Where the Finite Set $X$ Fits In**

Although $U$ and $V$ live in $\mathbb{R}^2$, we are interested in how they relate to the finite set $X$.

A natural topological structure on \(X\) comes from the **subspace topology**.

### Subspace topology on \(X\)
Define
$$
\tau_X=\{U\cap X : U\in \tau\}.
$$
Then $(X,\tau_X)$ is a topological space.

> **In words:**
> A subset $A\subseteq X$ is open in $X$ iff there exists an open set $O\subseteq \mathbb{R}^2$ such that  
> $$
 A = O\cap X.
  $$

So, from the plots:
- The open set in the topological space $(X,\tau_X)$ corresponding to $U$ is $U \cap X$.
- The open set corresponding to $V$ is $V \cap X$.

> **Interpretation:**
> - Points of $X$ that lie inside the shaded regions belong to $U \cap X$. (or $V \cap X$.).
> - Points outside do not.

## **6. Neighborhood Interpretation (Key Intuition)**

A point $x\in X$ is “interior” to an open region (like $U$ ) if:

$$
\exists r>0\ \text{such that}\ B(x,r)\subseteq U.
$$

This is the classical neighborhood definition of openness:

> A set is open if every point in it has a neighborhood contained in the set.

Our pictures demonstrate this idea visually: each open ball is literally drawn as a neighborhood region.


## **7. Why Two Frames?**

The two-frame figure is a didactic separation which means that $U,V$ both built from neighborhoods.


This reinforces the principle:

$$
U,\ V\ \text{are open} \quad \Longrightarrow \quad U\cup V\ \text{is open, and } U\cap V\ \text{is open.}
$$

## **Summary**

- We fixed a finite set $X\subset \mathbb{R}^2$.
- We constructed two open sets $ (U,V) \in \mathbb{R}^2$ as unions of open balls.
- This illustrates how the standard topology on $\mathbb{R}^2$ is generated by neighborhoods.
- Restricting to $X$  via the subspace topology gives open sets $U \cap X$ and $V \cap X$, which are the genuinely “topological” open sets in $(X,\tau_X)$.
