# üßÆ **A. Residue in Complex Analysis**

In complex analysis, the residue of a function at a point is a special number that captures how the function behaves near a singularity (a point where the function "blows up" or becomes undefined). Mathematically, it's the coefficient of the $\frac{1}{z-z_0}$ term in the Laurent series expansion of the function around that point $ùëß_0$. Residues help us evaluate complex integrals easily. If you integrate a function around a closed loop that encloses singularities, the result depends only on the residues at those points‚Äînot the full function. This is the heart of the Cauchy Residue Theorem.

>**Residue = the key number that tells us how a function behaves near a singularity, and it's what makes contour integration so powerful.**

## **1. Definition via Laurent Series**

Let $f(z)$ be a complex function that is analytic in a punctured neighborhood around a point $z_0$, i.e., $f(z)$ is analytic on

$$\{ z \in \mathbb{C} : 0 < |z - z_0| < r \}$$

Then $f(z)$ admits a Laurent series expansion about $z_0$:

$$f(z) = \sum_{n=-\infty}^{\infty} a_n (z - z_0)^n$$



The **residue** of $f$ at  $z_0$, denoted  $\operatorname{Res}(f, z_0)$, is defined as the coefficient of $(z - z_0)^{-1}$:

$$\operatorname{Res}(f, z_0) := a_{-1}$$

## **2. Pole in Complex Analysis**

In complex analysis, a **pole** is a type of singularity where a function becomes infinitely large in a controlled, algebraic way.

More precisely, a function $f(z)$ has a **pole of order $m$** at a point $z_0$  if near $ z_0$, it behaves like:


$$f(z) \sim \frac{1}{(z - z_0)^m}\quad \text{for some } m \in \mathbb{N}$$

This means that as $z \to z_0$, the function "blows up" but in a predictable fashion.

### **Classification**

- If $m = 1$, then $z_0$ is a **simple pole**.
- If $m > 1$, then $z_0$ is a **pole of order $m$**.

Let $f(z)$ be analytic in a punctured neighborhood around $z_0$, and suppose it admits a Laurent series expansion:

$$f(z) = \sum_{n=-m}^{\infty} a_n (z - z_0)^n \quad \text{with } a_{-m} \neq 0$$

Then $z_0$ is a pole of order $m$. The **principal part** of the Laurent series contains finitely many negative powers, up to $z - z_0)^{-m}$.

> **Unlike essential singularities (which behave wildly), poles are tame: they have a finite number of negative powers in their Laurent series. That‚Äôs why they‚Äôre so useful in residue calculus‚Äîbecause we can isolate and compute their contribution precisely.**


## **3. Methods of Computing Residues**

### (a) Simple Pole at $z_0$

If $f(z)$ has a simple pole at $z_0$, then:

$$ \operatorname{Res}(f, z_0) = \lim_{z \to z_0} (z - z_0) f(z)$$


### (b) Pole of Order $ m $

If $ f(z)$ has a pole of order $ m \geq 1$ at  $z_0$, then:

$$\operatorname{Res}(f, z_0) = \frac{1}{(m - 1)!} \lim_{z \to z_0} \frac{d^{m - 1}}{dz^{m - 1}} \left[ (z - z_0)^m f(z) \right]$$


### (c) Using Laurent Expansion

If the Laurent series is known explicitly, simply extract the coefficient of $(z - z_0)^{-1}$ .

### (d) Rational Function Form

If $ f(z) = \frac{g(z)}{h(z)}$, where:
- $ g $ and $ h $ are analytic near $z_0$,
- $ h(z_0) = 0 $, and
- $h'(z_0) \neq 0$,

Then $f$ has a simple pole at $z_0$, and:

$$\operatorname{Res}(f, z_0) = \frac{g(z_0)}{h'(z_0)}$$

## 4. Theoretical Significance

- **Linearity**:  
  
  $$\operatorname{Res}(af + bg, z_0) = a \operatorname{Res}(f, z_0) + b \operatorname{Res}(g, z_0)$$

- **Locality**:  
  The residue depends only on the behavior of $ f $ near $z_0$.

- **Topological Invariance**:  
  The value of the contour integral depends only on the residues inside the contour, not on the shape of the contour itself.

# **B. Singularity**

In complex analysis, a singularity is a point where a function stops being well-behaved‚Äîspecifically, where it‚Äôs no longer analytic.

Normally, a complex function is analytic if it‚Äôs smooth and differentiable in a neighborhood around a point. But at a singularity, something breaks: the function might blow up, become undefined, or behave unpredictably. Yet, it might still be analytic in a punctured neighborhood around that point (i.e., everywhere nearby except at the point itself). There are three main types of singularities:

* **Removable singularity:** The function misbehaves at the point, but you could ‚Äúfix‚Äù it by redefining the value there.

* **Pole:** The function goes to infinity in a controlled way, like $\frac{1}{z-z_0}.$
* **Essential singularity:** The function behaves wildly, with infinitely many negative powers in its Laurent series‚Äîlike $ùëí^{1/ùëß}$ near $ùëß=0$.


## Rigorous Definition of a Singularity

Let $f(z)$) be a complex-valued function defined on an open subset $ U \subset \mathbb{C}$, and let $z_0 \in U$ be a point such that, $f$ is **analytic** on a punctured neighborhood of $z_0$, i.e.,

$$f(z) \text{ is analytic on } \{ z \in \mathbb{C} : 0 < |z - z_0| < r \}
\quad \text{for some } r > 0,$$

but $f$ is **not analytic** at $z_0$ itself.

Then $z_0$ is called an **isolated singularity** of $f$.

## üß† Interpretation

An isolated singularity is a point where the function fails to be analytic, but is analytic everywhere else in a small punctured disk around that point. This allows us to study the behavior of $f(z)$ near $z_0$ using tools like Laurent series and classify the singularity as:

- **Removable**: if the Laurent series has no negative powers.
- **Pole**: if the Laurent series has finitely many negative powers.
- **Essential**: if the Laurent series has infinitely many negative powers.

# ‚ú≥Ô∏è Classification of Singularities in Complex Analysis

## 1. Isolated Singularities

Let $f(z)$ be a complex-valued function defined on an open subset $ U \subset \mathbb{C}$, and let $z_0 \in U$ be such that, $f$ is **analytic** on a punctured neighborhood of $z_0$, i.e.,

$$f(z) \text{ is analytic on } \{ z \in \mathbb{C} : 0 < |z - z_0| < r \}
\quad \text{for some } r > 0,$$

but $ f $ is **not analytic** at $z_0$ itself.

Then $z_0$ is called an **isolated singularity** of $f$.

## 2. Classification via Laurent Series

Suppose $f(z)$ admits a Laurent series expansion about $z_0$:

$$f(z) = \sum_{n=-\infty}^{\infty} a_n (z - z_0)^n
\quad \text{valid for } 0 < |z - z_0| < r$$

The nature of the singularity at $z_0$ is determined by the **principal part**, i.e., the terms with negative powers.

##üîπRemovable Singularity

If

$$ a_n = 0 \quad \forall n < 0,$$


then the singularity is **removable**. That is, $ f(z)$ can be redefined at  $z_0$  such that it becomes analytic on the full disk $ |z - z_0| < r $.

## Pole of Order $m$

If

$$
a_{-m} \neq 0 \quad \text{and} \quad a_n = 0 \text{ for all } n < -m,
$$


then $z_0$ is a **pole of order $m$**. The function diverges like $ \frac{1}{(z - z_0)^m}$ as $ z \to z_0 $.

- If $ m = 1$, it is a **simple pole**.
- If $ m > 1 $, it is a **higher-order pole**.

##üî∫Essential Singularity

If

$$a_n \neq 0 \quad \text{for infinitely many } n < 0,$$


then $z_0$ is an **essential singularity**. The behavior of $f(z)$ near $z_0$ is highly erratic and cannot be captured by a finite principal part.

## 3. Non-Isolated Singularities

A point $ z_0 \in \mathbb{C}$ is a **non-isolated singularity** of a function $f(z)$ if:

- There is **no punctured neighborhood** around $ z_0$ in which $ f $ is analytic, and
- Every neighborhood of $ z_0 $ contains **infinitely many singular points** of $f.$
In other words, the singularities **accumulate** at $z_0$, and $z_0$ is a **limit point** of singularities.
### üß† Example

Let

$$f(z) = \sum_{n=1}^{\infty} \frac{1}{z - n}$$


This function has singularities at every positive integer $ n \in \mathbb{N}.$ The point $z = \infty$ is a **non-isolated singularity**, because the singularities accumulate toward infinity.
