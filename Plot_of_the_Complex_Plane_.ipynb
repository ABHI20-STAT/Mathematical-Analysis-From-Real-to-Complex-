# **Plot each of the following sets in the complex plane.**
## 1.   $\mathfrak{S} = \Bigg\{ z \in \mathbb{C}: \frac{\pi}{6} <\arg z < \frac{\pi}{4}\Bigg\}$
## 2.   $\mathfrak{S} = \Bigg\{ z \in \mathbb{C}: 2 <|z| < 4 \Bigg\} $

---

1) Let $z= x+iy , \forall x,y \in \mathbb{R}.$ Write $z$ in polar form $z = re^{i\theta}; \;  r \geq 0$ and $\theta = \arg z$ (principal argument taken $\Big(\pi, - \pi\Big]).$. The set,

$$ \mathfrak{S} = \Bigg\{ z \in \mathbb{C}: \frac{\pi}{6} <\arg z < \frac{\pi}{4}\Bigg\}$$

is characterized as follows.

-   Since, $\arg 0$ is undefined, so $0 \notin \mathfrak{S}.$ Hence, $\forall z \in \mathfrak{S}$ satisfies $r>0.$

-   By definition

       $$\mathfrak{S} =\ re^{i\theta}; r>0, \frac{\pi}{6} <\theta < \frac{\pi}{4} $$

-   For $r>0$ and $\theta \in \Big(\frac{\pi}{6}, \frac{\pi}{4}\Big),$

    $$
    x = r \cos \theta , \quad y = r \sin \theta
    $$

Since, $\theta \in (0, \frac{\pi}{2})$ we have $\cos \theta >0$ and therefore $x>0.$

$$\implies \frac{y}{x} = \tan \theta$$ The tangent function is strictly increasing on $(0, \frac{\pi}{2}),$ so

$$
\Bigg\{\frac{\pi}{6} <\theta < \frac{\pi}{4}\Bigg\} \iff \tan\Big(\frac{\pi}{6}\Big) < \frac{y}{x} < \tan \Big(\frac{\pi}{4} \Big)
$$

$$
\implies  \tan\Big(\frac{\pi}{6}\Big) = \frac{1}{\sqrt{3}} , \quad \tan\Big(\frac{\pi}{4}\Big) =1
$$

Thus for $z\not=0,$

$$
\frac{\pi}{6} <\arg z < \frac{\pi}{4} \iff x>0 \quad \frac{1}{\sqrt{3}} < \frac{y}{x}<1
$$

Multiplying the double inequality by the positive number $x$ yields the region in the $(x,y)-$plane:

$$
\mathfrak{S}:= \Bigg\{ x+iy \in \mathbb{C}: x>0,  \frac{x}{\sqrt{3}} < y <1 \Bigg\}
$$

Therefore, Finally,

$$\boxed{\mathfrak{S} = \Bigg\{ re^{i\theta}: r>0, \frac{\pi}{6} <\theta < \frac{\pi}{4}\Bigg\} = \Bigg\{ x+iy \in \mathbb{C}: x>0,  \frac{x}{\sqrt{3}} < y <1 \Bigg\}}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Define sector parameters ---
R = 6.0
theta1, theta2 = np.pi/6, np.pi/4

# --- Compute coordinates for the filled region ---
theta = np.linspace(theta1, theta2, 400)
x_poly = np.concatenate(([0], R*np.cos(theta), [0]))
y_poly = np.concatenate(([0], R*np.sin(theta), [0]))

# --- Prepare figure ---
fig, ax = plt.subplots(figsize=(8, 8))

# Fill the open sector region (visual representation of S)
ax.fill(x_poly, y_poly, alpha=0.25, label=r'$\mathfrak{S} = \{re^{i\theta}: r>0, \pi/6<\theta<\pi/4\}$')

# --- Draw the two bounding rays (not included, dashed) ---
ax.plot([0, R*np.cos(theta1)], [0, R*np.sin(theta1)], linestyle='--', linewidth=1.2, label=r'$\theta = \pi/6$ (ray not included)')
ax.plot([0, R*np.cos(theta2)], [0, R*np.sin(theta2)], linestyle='--', linewidth=1.2, label=r'$\theta = \pi/4$ (ray not included)')

# --- Cartesian interpretation lines (y = (1/sqrt(3))x and y = x) ---
x_line = np.linspace(0, R, 300)
ax.plot(x_line, (1/np.sqrt(3))*x_line, linestyle=':', linewidth=1.0, label=r'$y = \frac{x}{\sqrt{3}}$')
ax.plot(x_line, x_line, linestyle=':', linewidth=1.0, label=r'$y = x$')

# --- Annotate key angles and regions ---
ax.text(2.5*np.cos((theta1+theta2)/2), 2.5*np.sin((theta1+theta2)/2),
        r'$\frac{\pi}{6} < \arg z < \frac{\pi}{4}$',
        fontsize=13, ha='center', va='center', bbox=dict(facecolor='white', alpha=0.6, edgecolor='none'))

# --- Draw axes ---
ax.axhline(0, color='black', linewidth=1.0)
ax.axvline(0, color='black', linewidth=1.0)

# --- Label axes ---
ax.set_xlabel(r'$\Re(z) = x$', fontsize=13)
ax.set_ylabel(r'$\Im(z) = y$', fontsize=13)

# --- Title and origin mark ---
ax.plot(0, 0, 'ko', markersize=4)
ax.text(0.1, -0.3, 'Origin (excluded)', fontsize=10, color='black')
ax.set_title(r'Visualization of $\mathfrak{S} = \{ z\in\mathbb{C} : \frac{\pi}{6}<\arg z<\frac{\pi}{4}\}$', fontsize=14, pad=15)

# --- Add legend and style ---
ax.legend(fontsize=9, loc='upper left', framealpha=0.95)
ax.set_xlim(-1, R)
ax.set_ylim(-1, R)
ax.set_aspect('equal', 'box')
ax.grid(True, linestyle=':', linewidth=0.7)

plt.show()
```

The above plot visualizes the mathematical deduction:

* The filled region represents

$$\mathfrak{S} = \Bigg\{ re^{i\theta}: r>0, \frac{\pi}{6} <\theta < \frac{\pi}{4}\Bigg\} = \Bigg\{ x+iy \in \mathbb{C}: x>0,  \frac{x}{\sqrt{3}} < y <1 \Bigg\}$$

* The dashed rays at angles $=/6$ and $=/4$ are the excluded boundaries of the sector.

* The dotted Cartesian lines $y=x/\sqrt{3}$ and $y=x$ correspond to the inequalities derived in the Cartesian form.

* The origin $(0, 0)$ is excluded since $ \arg(0)$ is undefined.

b\) Let $z= x+iy , \forall x,y \in \mathbb{R}.$ Denote the complex modulus by $|z| = \sqrt{x^2+y^2}.$

Now, consider the set,

$$\mathfrak{S} = \Bigg\{ z \in \mathbb{C}: 2 <|z| < 4 \Bigg\}$$ Now, we express $z=re^{i\theta}$ with $z = re^{i\theta}; \;  r \geq 0, \forall \theta \in [0,2\pi].$ Then

$$\mathfrak{S} = \Bigg\{  re^{i\theta}: 2 < r < 4, \theta \in [0,2\pi) \Bigg\}$$

That is, $\mathfrak{S}$ is the open ring with inner radius $2$ & outer radius $4.$

The polar inequality $2< |z| <4$ is equivalent to

$$
2< \sqrt{x^2+y^2} <4 \quad \iff \quad 4<x^2+y^2<16
$$

Thus, $$
\boxed{\mathfrak{S}: = \Bigg\{ x+iy\in \mathbb{C}: 4 < x^2+y^2 < 16 \Bigg\}}
$$

Therefore, the final answer is,

$$\mathfrak{S} = \Bigg\{ z \in \mathbb{C}: 2 <|z| < 4 \Bigg\} = \Bigg\{  re^{i\theta}: 2 < r < 4, \theta \in [0,2\pi) \Bigg\} = \Bigg\{ x+iy\in \mathbb{C}: 4 < x^2+y^2 < 16 \Bigg\}$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Define polar grid for the annulus
theta = np.linspace(0, 2*np.pi, 800)
r_inner = 2
r_outer = 4

# Compute coordinates
x_inner, y_inner = r_inner * np.cos(theta), r_inner * np.sin(theta)
x_outer, y_outer = r_outer * np.cos(theta), r_outer * np.sin(theta)

# Create figure
fig, ax = plt.subplots(figsize=(9, 9))
ax.set_facecolor("black")

# Fill the annulus region
theta_fill = np.linspace(0, 2*np.pi, 600)
r_vals = np.linspace(r_inner, r_outer, 100)
R, T = np.meshgrid(r_vals, theta_fill)
X, Y = R * np.cos(T), R * np.sin(T)
ax.pcolormesh(X, Y, np.ones_like(R), cmap='viridis', shading='auto', alpha=0.8)

# Boundary circles
ax.plot(x_inner, y_inner, color='white', lw=2.5, label=r"$|z|=2$ (Inner Boundary)")
ax.plot(x_outer, y_outer, color='cyan', lw=2.5, label=r"$|z|=4$ (Outer Boundary)")

# Annotate main geometric features
ax.text(0, 0, r"Hole at origin", color='white', ha='center', va='center', fontsize=12, style='italic')
ax.text(3*np.cos(np.pi/4), 3*np.sin(np.pi/4), r"$\mathfrak{S}$: Open~Annulus", color='white', fontsize=13, fontweight='bold')

# Mark center and radii lines
ax.scatter(0, 0, color='red', s=40, label=r"Origin $(0,0)$")
for angle, color in zip([0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi], ['gray']*5):
    ax.plot([0, r_outer*np.cos(angle)], [0, r_outer*np.sin(angle)], color=color, lw=0.8, ls='--', alpha=0.7)

# Axis formatting
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_aspect('equal', 'box')
ax.set_xlabel(r'$\Re(z)$', fontsize=14, color='white')
ax.set_ylabel(r'$\Im(z)$', fontsize=14, color='white')
ax.set_title(r'$\mathfrak{S} = \{ z \in \mathbb{C} : 2 < |z| < 4 \}$  Rigorous Geometric Visualization', fontsize=15, color='white', pad=20)

# Grid and legend
ax.grid(True, color='gray', linestyle=':', linewidth=0.5, alpha=0.5)
ax.legend(facecolor="black", edgecolor="white", labelcolor='white', fontsize=10)

# Style tweaks
for spine in ax.spines.values():
    spine.set_edgecolor('white')

plt.show()
```

The above visualization precisely illustrates the mathematical structure of

$$\mathfrak{S}=\{zC:2<z<4\}.$$

Interpretation linked to deduction:

* The shaded annular region represents all complex numbers $z=re^{i}$ with modulus $r$ strictly between $2$ and $4$.

* The two circular boundaries, $z=2$ (inner, white) and $z=4$ (outer, cyan), are not included, reflecting the strict inequalities.

* The origin is marked and explicitly excluded  points with $z2$ lie outside the set.
